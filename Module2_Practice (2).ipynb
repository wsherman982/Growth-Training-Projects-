Assume a game in which two players, Player A and Player B, roll a six-sided die. The game continues until one of the players reaches a score of 50 or more. The player with the highest score at the end of the game wins. If both players reach 50 or more in the same round, the player with the higher score wins. If both players have the same score, the game is a tie. We will write a program that simulates this game and keeps track of the scores.

```python
import random

def roll_die():
    '''Simulates rolling a six-sided die and returns the result (1-6).'''
    return random.randint(1,6)
print("Rolled:", roll_die())
```

```python
def play_game(target_score=50):
     '''Runs game until player A or Player B reaches 50'''


     a_score = 0
     b_score = 0
     round_num = 1

     while True: #Input loop
         a_roll = roll_die()
         b_roll = roll_die()

         #Adds rolls to scores
         a_score += a_roll
         b_score += b_roll
         
         print(f"Round {round_num}: A rolled {a_roll} (total {a_score}), "f" B rolled {b_roll} (total {b_score})")

         #check winner

         if a_score >= target_score or b_score >= target_score:
             print("\nWe Have a Winnah")
             print(f"Final Score: A = {a_score}, B = {b_score}")

             if a_score > b_score:
                 print("Player A WINS!!! As they should!")
             elif b_score > a_score:
                 print("Player B WINS, We love a B team!!!")
             else:
                 print("It's a tie! Wow the crowd is going WILDDD!!")
             break



         round_num += 1

play_game(50)
```
